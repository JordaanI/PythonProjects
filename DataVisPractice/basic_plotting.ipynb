```python
###############
### Imports ###
###############

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
#####################
### Generate data ###
#####################

### X_SPACE
X = np.linspace(-10,10,200)

### Parameters
rng = np.random.default_rng()
slopes = rng.integers(low=-10, high=10, size=3)
intercepts = rng.integers(low=-10, high=10, size=3)

### Colour & markers
colours = ['g','r','y']
markers = ['^','*','+']
```

```python
##################
### Plot Lines ###
##################

fig, ax = plt.subplots()
for i in range(3):
    colour = colours[i]
    marker = markers[i]
    current_label = f'Slope: {slopes[i]}, Intercept: {intercepts[i]}'
    ax.plot(X, slopes[i]*X+intercepts[i], label=current_label, color=colour, marker=marker)
ax.legend()
plt.show()
```

```python
#################
### SUB_PLOTS ###
#################

rows = 1
columns = len(slopes)

fig, axes = plt.subplots(rows,columns, figsize=[12,6])
for i in range(3):
    colour = colours[i]
    marker = markers[i]
    current_label = f'Slope: {slopes[i]}, Intercept: {intercepts[i]}'
    axes[i].plot(X, slopes[i]*X+intercepts[i], label=current_label, color=colour, marker=marker)
    axes[i].legend()
plt.show()
```

```python
###############
### Seaborn ###
###############

tips = sns.load_dataset('tips')

fig, ax = plt.subplots(figsize=(12,6))
sns.barplot(x='sex', y='tip', data=tips)
plt.show()
```
