```python
import random

class vehicle:
    """Generates a vehicle class"""
    def __init__(self,name) -> None:
        #The name of the vehicle
        self.name = name
    def query(self):
        return f'The vehicle name is {self.name}'

class car(vehicle):
    """Generates a car subclass"""
    def __init__(self, name, price=0) -> None:
        super().__init__(name)
        self._make = random.choice(['Mazda','Toyota','Hyundai','Opel','Citroen'])
        self._model = random.choice(list(range(2000,2020,1)))
        self._price = price
    
    @property
    def make(self):
            return self._make

    @make.setter
    def make(self, make):
        if make not in ['Mazda','Toyota','Hyundai','Opel','Citroen']:
            raise ValueError('Not accepted vehicle type')
        else:
            self._make = make

    @property
    def model(self):
        return self._model
    
    @model.setter
    def model(self, model):
        if model not in list(range(2000,2020,1)):
            raise ValueError('Not accepted model number')
        else:
            self._model = model

    @property
    def price(self):
        return self._price
    
    @price.setter
    def price(self, price):
        self._price = price

    def query(self):
        return super().query() + f', the make is {self._make}, the model is {self._model} and the price is {self._price}'
```

```python
Toyota = car('Princess')
print(Toyota.query())
Toyota.price = 15000
print(Toyota.query())
Toyota.make = 'Toyota'
Toyota.model = 2019
print(Toyota.query())
```

```python
class business:
    def __init__(self,balance = 1000000, credit = 0, debit = 0) -> None:
        self._balance = balance
        self._credit = credit
        self._debit = debit
    
    @property
    def balance(self):
        return self._balance
    
    @balance.setter
    def balance(self,balance):
        self._balance = balance
    
    @property
    def credit(self):
        return self._credit
    
    @credit.setter
    def credit(self,credit):
        self._credit = credit

    @property
    def debit(self):
        return self._debit
    
    @debit.setter
    def debit(self,debit):
        self._debit = debit

    def equalize_funds(self):
        self._balance += self._credit
        self.balance -= self._debit
    
    def cash_sale(self, customer, amount):
        self._balance += amount
        customer._sales += amount

    def credit_sale(self, customer, amount):
        self._credit += amount
        customer.debit += amount

class customer():
    def __init__(self,name, sales = 0, debit = 0, credit = 0) -> None:
        self._name = name
        self._sales = sales
        self._debit = debit
        self._credit = credit
#Setters and getters
    @property
    def name(self):
        return self._name
    
    @name.setter
    def name(self,name):
        self._name = name

    @property
    def debit(self):
        return self._debit
    
    @debit.setter
    def debit(self,debit):
        self._debit = debit

    @property
    def credit(self):
        return self._credit
    
    @credit.setter
    def credit(self,credit):
        self._credit = credit

    @property
    def sales(self):
        return self._sales
    
    @sales.setter
    def sales(self,sales):
        self._sales = sales
# Credit and debit methods
    def credit_funds(self,business,amount):
        business.debit += amount
        self._credit += amount
        return f'Business debitted {amount} and customer {self._name} creditted to {self._credit}'

    def debit_funds(self,business,amount):
        business.credit += amount
        self._debit += amount
        return f'Business debitted {amount} and customer {self._name} debitted to {self.debit}'
```
